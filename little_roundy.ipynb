```python
def url_open(url):
    req = urllib.request.Request(url)
    req.add_header("User-Agent","Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36")
    response = urllib.request.urlopen(req)
    html = response.read()

    return html
```

### 一、抓取图片网页的url和name

```python
# 遍历index（pageurl）页面
pageurl_list = []

def PageUrl(num): #左闭右开
    for i in range(1,num):
        if i == 1:#第一个网页
            pageurl = 'https://www.chinabuddhism.com.cn/tp/index.html'
            pageurl_list.append(pageurl)
        else:
            pageurl = baseurl + '_' + str(i) +'.html'
            pageurl_list.append(pageurl)
        # print(pageurl)
    return pageurl_list
```

```python
# 从index网页中抓取href和name
href_list = []
name_list = []

def getWebUrl(pageurl):

    #打开pageurl网页
   page = url_open(pageurl).decode('gbk','ignore') #‘document.charset’查看网页encode方式'gb2312'
   page.encode(encoding='utf-8')

   #抓取网页中的图片网址
   soup = BeautifulSoup(page,'html.parser')
   # print(soup.prettify())
   for link in soup.find_all('li',id="pic_title_c"):
       a = link.next_element
       # print(type(a))
       href = a.get('href') #获取图片的网页地址
       href_list.append(href)
       name = a.next_element #获取图片名称
       name_list.append(name)

   return href_list  #href'/tp/2017-05-19/5204.html'
   return name_list
```

```python
#抓取index页面里的href和name
num = 243
PageUrl(num)
print(len(pageurl_list))
```

```python
for i in range(0,len(pageurl_list)):
    pageurl = pageurl_list[i]
    try:
        getWebUrl(pageurl)
    except Exception as e:
     print(pageurl)
     print(e)
    i = i+1
```

```python
# 转译href为image url
weburl_list = []
for i in range(0,len(href_list)):
    if 'chinabuddhism' in href_list[i]:
      weburl = href_list[i]
      weburl_list.append(weburl)
        
    else:
      weburl = 'https://www.chinabuddhism.com.cn'+ href_list[i]
      weburl_list.append(weburl)
       
        
    i = i+1
print(len(weburl_list))
```

### 二、抓取imageurl里面的大图

```python
# 抓取weburl中的imgurl,画在align="center"里面
def getImgUrl_1(weburl):
    web = url_open(weburl).decode('gbk')
    web.encode(encoding='utf-8')

    websoup = BeautifulSoup(web,'html.parser')
    for element in websoup.find_all('div',align="center"):
        element_img = element.findNext('img')
        imgurl = element_img.get('src')
        # print(imgurl)
    return imgurl
```

```python
# 抓取weburl中的imgurl，画在class="content_zi"里
def getImgUrl_2(weburl):
    web = url_open(weburl).decode('gbk')
    web.encode(encoding='utf-8')

    websoup = BeautifulSoup(web,'html.parser')
    for element in websoup.find_all('div',class_="content_zi"):
        element_img = element.findNext('img')
        imgurl = element_img.get('src')
        # print(imgurl)
    return imgurl
```
